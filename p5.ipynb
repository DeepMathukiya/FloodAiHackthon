```python
# Open the GRIB file
import pygrib
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx

grbs = pygrib.open('data1.grib')


# Assuming you identified the rainfall message with the correct name
rainfall_grb = grbs.select(name='Total precipitation')[0]  # Replace with the correct name

# Extract data
rainfall_data = rainfall_grb.values
latitudes, longitudes = rainfall_grb.latlons()

# Close the GRIB file
grbs.close()

# Define the latitude and longitude bounds for Mumbai
lat_min, lat_max = 18.75, 19.25
lon_min, lon_max = 72.75, 73.25

# Create masks to filter data for Mumbai region
lat_mask = (latitudes >= lat_min) & (latitudes <= lat_max)
lon_mask = (longitudes >= lon_min) & (longitudes <= lon_max)

# Filter latitude and longitude for the selected region
filtered_latitudes = latitudes[lat_mask]
filtered_longitudes = longitudes[lon_mask]

# Create a meshgrid for latitudes and longitudes
lon_grid, lat_grid = np.meshgrid(filtered_longitudes, filtered_latitudes)

# Extract indices where the masks are True
lat_indices = np.where(lat_mask)[0]
lon_indices = np.where(lon_mask)[0]

# Filter the rainfall data using the latitude and longitude indices
filtered_rainfall = rainfall_data[np.ix_(lat_indices, lon_indices)]

# Convert the filtered data to a GeoDataFrame
# Create a DataFrame with columns for latitude, longitude, and precipitation
data = {
    'longitude': lon_grid.flatten(),
    'latitude': lat_grid.flatten(),
    'precipitation': filtered_rainfall.flatten()
}
df = pd.DataFrame(data)

# Convert DataFrame to GeoDataFrame
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.longitude, df.latitude))
gdf.set_crs(epsg=4326, inplace=True)  # Set the coordinate reference system to WGS84

# Plotting with a basemap
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax, markersize=10, c='precipitation', cmap='Blues', legend=True, alpha=0.7)
ctx.add_basemap(ax, crs=gdf.crs.to_string(), source=ctx.providers.CartoDB.Positron)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Precipitation Data for Mumbai Region')
plt.show()
```
